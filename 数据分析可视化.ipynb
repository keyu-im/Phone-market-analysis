```python
import pandas as pd
import numpy as np
```

```python
#载入数据
#每1000行为一个数据模块，逐块读取
df=pd.read_csv('data.csv')
chunker=pd.read_csv('data.csv',chunksize=1000)
for i in chunker:
  print(i)
df.head(5)
```

```python
#更改中文表头
df.columns=['序号','手机名字','价格','手机卡','处理器','内存','电池','尺寸','像素','型号']
df=pd.read_csv('data.csv',names=['序号','手机名字','价格','手机卡','处理器','内存','电池','尺寸','像素','型号'])
df=df.drop(index=0)
df=df.reset_index(drop=True)
df.head(3)
```

```python
#处理缺失值
df.info()
misscode=df[df['型号'].isnull()]
print(misscode[['手机名字','价格']])
#这条不完整数据的其他部分信息完整，且型号为次要分析变量，所以数据保留
df['型号']=df['型号'].fillna('Unknown')
```

```python
#将'价格'的数据清洗为数字
df['价格']=df['价格'].str.replace('?','',regex=False)
df['价格']=df['价格'].str.replace(',','',regex=False)
df['Price']=df['价格'].astype(float)
```

```python
#将内存分为'RAM'和'inbuilt'并转换为float
df['RAM_demo']=df['内存'].str.split(',').str[0]
df['RAM']=df['RAM_demo'].str.extract(r'(\d+\.?\d*)')
df['RAM']=df['RAM'].astype(float)
df['inbuilt_demo']=df['内存'].str.split(',').str[1]
df['inbuilt']=df['inbuilt_demo'].str.extract(r'(\d+\.?\d*)')
df['inbuilt']=df['inbuilt'].astype(float)
df=df.drop(columns=['RAM_demo','inbuilt_demo'])
```

```python
#将手机屏幕尺寸转换为float
df['尺寸_']=df['尺寸'].str.split(',').str[0]
df['Screensize_inches']=df['尺寸_'].str.extract(r'(\d+\.?\d*)')
df['Screensize_inches']=df['Screensize_inches'].astype(float)
df=df.drop(columns=['尺寸_'])
```

```python
#将电池转换为float
df['Battery_capacity_mAh']=df['电池'].str.extract(r'(\d+)')
df['Battery_capacity_mAh']=df['Battery_capacity_mAh'].astype(float)
```

```python
#清洗手机品牌名
df['Brand_name']=df['手机名字'].str.split().str[0]
#清洗处理器名字
df['Processor']=df['处理器'].str.split(',').str[0]
```

```python
df.info()
#处理缺失值，删除'RAM','inbuilt','Screensize_inches'三个关键分析变量的缺失值
#'Battery_capacity'作为非关键分析变量，将空缺补充为中位数
df=df.dropna(subset=['RAM','inbuilt','Screensize_inches'])
df=df.reset_index(drop=True)
Battery_median=df['Battery_capacity_mAh'].median()
df['Battery_capacity_mAh']=df['Battery_capacity_mAh'].fillna(Battery_median)
```

```python
#处理重复值
df=df.drop_duplicates()
df=df.reset_index(drop=True)
```

```python
#处理异常值,删除不合理的极值
print(df[['Price','RAM','inbuilt','Battery_capacity_mAh','Screensize_inches']].describe())
df=df.drop(df[df['RAM']==df['RAM'].max()].index)
df=df.drop(df[df['Screensize_inches']==df['Screensize_inches'].max()].index)
df=df.drop(df[df['Screensize_inches']==df['Screensize_inches'].min()].index)
df=df.drop(df[df['Battery_capacity_mAh']==df['Battery_capacity_mAh'].min()].index)
#查看'RAM'箱型图以及处理异常值
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.boxplot(df['RAM'])
plt.title('RAM boxplot')
plt.show()
top=df['RAM'].quantile(0.75)+1.5*(df['RAM'].quantile(0.75)-df['RAM'].quantile(0.25))
df=df[df['RAM']<=top]
print(df['RAM'].describe())
```

```python
#查看'Screensize_inches'箱型图
plt.figure(figsize=(8,5))
plt.boxplot(df['Screensize_inches'])
plt.title('Screensize boxplot')
plt.show()
#清理screensize异常值
df=df[df['Screensize_inches']>=3]
print(df['Screensize_inches'].describe())
```

```python
#查看'inbuilt'箱型图
plt.figure(figsize=(8,5))
plt.boxplot(df['inbuilt'])
plt.title('inbuilt boxplot')
plt.show()
#剔除极端值
df=df[df['inbuilt']<400]
print(df['inbuilt'].describe())
```

```python
df=df.reset_index(drop=True)
#检查'Brand_name'和'Processor'，处理数据一致性
print(df['Brand_name'].unique())
print(df['Brand_name'].value_counts())
print(df['Processor'].unique())
print(df['Processor'].value_counts())
```

```python
#价格分布图
import seaborn as sns
plt.figure(figsize=(6,5))
sns.set_theme(style="darkgrid",palette="pastel")
sns.histplot(x='Price',kde=True,data=df)
plt.title('Price distribution')
plt.show()
```

```python
#品牌市场占有率
plt.figure(figsize=(6,5))
sns.countplot(data=df, x='Brand_name', order=df['Brand_name'].value_counts().head(25).index, palette="pastel")
plt.title("Market share")
plt.xticks(rotation=45,fontsize=8)
plt.ylabel("Count")
plt.xlabel('Brand name')
plt.show()
#由图可知Realme，Samsung，Xiaomi，Vivo占据市场的主要份额
```

```python
#品牌的价格分布和对应市场份额
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Brand_name',order=df['Brand_name'].value_counts().head(25).index,y='Price')
plt.xticks(rotation=90)
plt.title('Brands price distribution')
plt.xlabel('Brand name')
plt.ylabel('Price')
plt.show()
#由图可知：
#0-25000INR为下沉市场，大部分品牌集中在该区间，占据市场的最大份额。说明消费者对中低价位手机需求旺盛
#25000-50000INR为中端市场，仅有少部分品牌属于该区间，消费者需求量有限
#50000INR+为高端市场，基本由Apple独占覆盖了顶级价格区间，且其产品价格跨度最广，横跨多个市场层级。
#结论：此地域的手机市场适合采用低价策略获取销量，但Apple的全价位的布局也获得了差异性优势
```

```python
#分析各品牌RAM的分布情况
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x='Brand_name',order=df['Brand_name'].value_counts().head(25).index,y='RAM')
plt.xticks(rotation=90)
plt.title('Brands RAM distribution')
plt.xlabel('Brand name')
plt.ylabel('RAM')
plt.show()
#4－8GB的中等配置占绝大部分市场；其次是少量8－12GB的中高端配置，以IQOO和OnePlus为主。
```

```python
#分析各品牌inbuilt的分布情况
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x='Brand_name',order=df['Brand_name'].value_counts().head(25).index,y='inbuilt')
plt.xticks(rotation=90)
plt.title('Brands inbuilt distribution')
plt.xlabel('Brand name')
plt.ylabel('inbuilt')
plt.show()
#由RAM箱体图和inbuilt箱体图可得知该地区的手机市场，占据第一梯队的为50－150GB的内存；其次是中等梯队的＞150GB的高端配置，以Apple,iQOO和OnePlus为代表；仅有少量低端配置的市场份额。
```

```python
#分析价格和其他变量的相关性热力图
items=['Price','Battery_capacity_mAh','RAM','inbuilt','Screensize_inches']
corr=df[items].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap='coolwarm',center=0)
plt.title('Correlation Heatmap')
plt.show()
#RAM、inbuilt与价格呈中等正相关，内存和储存较大的手机价格越高，说明配置方面是手机产家的主要溢价领域。而电池容量，屏幕大小则与价格关联性较弱，表明厂家在这些方面有更多灵活调整的空间。
```
